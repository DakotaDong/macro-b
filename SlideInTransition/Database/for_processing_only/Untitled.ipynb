```python
import pandas as pd
```

```python
family_df = pd.read_csv('family.csv', names=['id','orderId','commonName','scientificName','pollution','description'],sep = ',', header=None)
```

```python
family_ref_df = pd.read_csv('families_ref.csv',sep = ',', header=0)
family_ref_df.head()
```

```python
family_ref_familyid_to_sciname = {}

for index, row in family_ref_df.iterrows():
    family_id = row['id']
    scientific_name = row['scientificName']
    family_ref_familyid_to_sciname[str(family_id)] = scientific_name
```

```python
family_ref_familyid_to_sciname
```

```python
family_df.head()
```

```python
family_sciname_to_correct_id = {}

for index, row in family_df.iterrows():
    scientific_name = row['scientificName']
    correct_id = row['id']
    family_sciname_to_correct_id[scientific_name] = str(correct_id)
```

```python
family_sciname_to_correct_id
```

```python
diag_df = pd.read_csv('family_diag_characters.csv', sep = ',', header=0)

#'relationshipTypeCode','charId','taxId','diagCharType','name','description','audio','image'
```

```python
diag_df.head()
```

```python
correct_ids = []
for index, row in diag_df.iterrows():
    ref_id = str(row['taxaObjectId'])
    family_name = family_ref_familyid_to_sciname[ref_id]
    correct_id = family_sciname_to_correct_id[family_name]
    correct_ids.append(correct_id)
```

```python
correct_ids
```

```python
result = pandas.DataFrame(data={"correct_id": correct_ids})
result.to_csv("correct_fam_id.csv", sep=',',index=False)
```
